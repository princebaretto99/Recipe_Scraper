```python
data = pd.read_csv("final_csv.csv")
len(data)
```

```python
data = data.iloc[:,2:]
```

```python
temp_data = data
temp_data.head(2)
```

```python
cal,sod,pro,fat,chol,carbo =[],[],[],[],[],[]
for i in range(len(temp_data["nutrition_text"])):
    string = str(temp_data["nutrition_text"][i])
    arr = string.split(";")
    
    calflag = False
    sodflag = False
    proflag = False
    cholflag = False
    carboflag = False
    fatflag = False
    
    for i in range(len(arr)):
        arr[i] = arr[i].strip()
        if(i==0):
            cal.append(arr[i].split(" ")[0])
            calflag = True
        else:
            if(arr[i].split(" ")[0] == "cholesterol"):
                cholflag = True
                chol.append(arr[i].split(" ")[1])
            if(arr[i].split(" ")[0] == "carbohydrates"):
                carboflag = True
                carbo.append(arr[i].split(" ")[1])
            if(arr[i].split(" ")[0] == "protein"):
                proflag= True
                pro.append(arr[i].split(" ")[1])
            if(arr[i].split(" ")[0] == "fat"):
                fatflag = True
                fat.append(arr[i].split(" ")[1])
            if(arr[i].split(" ")[0] == "sodium"):
                sodflag = True
                sod.append(arr[i].split(" ")[1])
    if(calflag == False):
        cal.append("0")
    if(carboflag == False): 
        carbo.append("0")
    if(fatflag == False):
        fat.append("0")
    if(proflag == False):
        pro.append("0")
    if(cholflag == False):
        chol.append("0")
    if(sodflag == False):
        sod.append("0")
```
